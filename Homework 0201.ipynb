### Question 4 [10 points]

- Initiate the logistic regression classifier from the sklearn library as shown below.
- Fit the model
- Make predictions
- Calculate accuracy score. What are the **training** and **testing** accuracies of the model?
- Create confusion matrix
- Either calculate using the confusion matrix, or use the methods in `metrics` to get the following metrics on the testing set:
     - Recall (Sensitivity)
     - Specificity
     - Precision
     - False Positive Rate
     - F1 Score

```python
from sklearn.linear_model import LogisticRegression

# initialize the Logistic Regression classifier
clf = LogisticRegression(max_iter = 10000, C = 0.1)

# fit the model to the training data
clf.fit(X_train, y_train)

# make predictions on the train set
y_pred_train = clf.predict(X_train)

# make predictions on the test set
y_pred_test = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# calculate accuracy score
accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print("Training Accuracy: ", accuracy_train)
print("Testing Accuracy: ", accuracy_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

# creating confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_test, cmap = "Blues")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score

conf_matrix = confusion_matrix(y_test, y_pred_test)
trueNeg, falsePos, falseNeg, truePos = conf_matrix.ravel()

# calculate recall
recall = recall_score(y_test, y_pred_test)

# calculate Specificity
specificity = trueNeg / (trueNeg + falsePos)

# calculate precision
precision = precision_score(y_test, y_pred_test)

# claculate false positive rate
false_positive_rate = falsePos / (falsePos + trueNeg)

# calculate f1 score
f1 = f1_score(y_test, y_pred_test)

print("Recall: ", recall)
print("Specificity: ", specificity)
print("Precision: ", precision)
print("False Positive Rate: ", false_positive_rate)
print("F1 Score: ", f1)
```

### Question 6 [10 points]

In logistic regression keeping the other parameters constant try these values for C: `C_list = [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 2, 10, 50, 100, 200, 500, 1000, 2000, 5000, 10000]`.  
C is the inverse of the regularization parameter $\lambda$. As C increases, the penalty decreases.
- Create a loop and fit the models using each C value. 
- Make predictions
- Calculate accuracy scores for **training** and **testing** datasets. Create a list of accuracy results for train and test accuracies. Show these lists.
- Create the validation curve based on each C value and the corresponding train and test accuracies.

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

C_list = [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 
          1, 2, 10, 50, 100, 200, 500, 1000, 2000, 5000, 10000]

# initialize empty lists to store performance metrics
train_accuracy_list = []
test_accuracy_list = []

# loop through each value of C
for c in C_list:
    
    # create logistic regression model with current value of C
    clf = LogisticRegression(max_iter = 5000, C = c)
    
    # fit the model on training set
    clf.fit(X_train, y_train)
    
    # make predictions on training set
    y_train_pred = clf.predict(X_train)
    
    # make predictions on testing set
    y_test_pred = clf.predict(X_test)
    
    # calculate accuracy for training set
    train_accuracy = accuracy_score(y_train, y_train_pred)
    
    # calculate accuracy for testing set
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    # append accuracy values to respective lists
    train_accuracy_list.append(train_accuracy)
    test_accuracy_list.append(test_accuracy)

# convert results to numpy arrays
train_accuracy_list = np.array(train_accuracy_list)
test_accuracy_list = np.array(test_accuracy_list)

# round the results to 3 decimal points
train_accuracy_list = np.round(train_accuracy_list, 3)
test_accuracy_list = np.round(test_accuracy_list, 3)

# print the results
for i in range(len(C_list)):
    print(f"C: {C_list[i]}, Train Accuracy: {train_accuracy_list[i]}, Test Accuracy: {test_accuracy_list[i]}")
```

```python
import matplotlib.pyplot as plt

# create a validation curve
f, ax = plt.subplots(dpi = 150)
plt.plot(train_accuracy_list, marker = 'o', color = 'blue', label = 'Train')
plt.plot(test_accuracy_list, color = 'orange', marker = 'o', label = 'Test')
plt.xticks(ticks = np.arange(len(C_list)), labels = C_list, rotation = 90)
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

### Question 9 [10 points]

- After choosing your best C value, let's check the learning curve.
- Create a logistic regression model of your chosen C value.
- Starting from 10 data points in the training set, increase your training set size by 1 points in each iteration, fit the model, get the accuracy scores for both the current training and test sets.
- Increase the training set size consecutively. Do not randomly select data points. (`X_train_ss[0:N,:] where N = {1,2,3,...,len(X_train_ss)}`)
- Plot the learning curve. What do you see? Comment on your finding. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# initialize empty lists to store performance metrics
accuracy_train = []
accuracy_test = []

# loop over different training set sizes
for n in range(10, X_train_ss.shape[0]):
    
    # initialize the Logistic Regression classifier
    clf = LogisticRegression(max_iter = 100, solver = 'lbfgs', penalty = 'l2', C = 0.5, verbose = 0)
    
    # fit the model to the first 'n' samples
    clf.fit(X_train_ss[0:n,:], y_train[0:n])


    # make predictions on the first 'n' training samples
    pred_train = clf.predict(X_train_ss[0:n,:])

    # make predictions on the first 'n' testing samples
    pred_test = clf.predict(X_test_ss)
    
    # append accuracy values to respective lists
    accuracy_train.append(metrics.accuracy_score(y_true = y_train[0:n], y_pred = pred_train))   
    accuracy_test.append(metrics.accuracy_score(y_true = y_test, y_pred = pred_test))
```

```python
import matplotlib.pyplot as plt

# define the range of number of training samples
n_range = range(10, X_train_ss.shape[0])

# create a learning curve
f, ax = plt.subplots(dpi = 150)
ax.plot(n_range, accuracy_train, color = 'blue', label = 'Train')
ax.plot(n_range, accuracy_test, color = 'orange', label = 'Test')
ax.set_xlabel('Number of points')
ax.set_ylabel('Accuracy')
ax.set_title("Learning Curve")
ax.legend()
ax.grid()
```

I am going to use ensemble and feature selection for this question.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score

# initialize the Logistic Regression classifier
lg_clf = LogisticRegression(C = 0.5)

# initialize the Random Forest classifier
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 42)

# generate voting ensemble
ensemble_model = VotingClassifier(estimators = [('logistic', lg_clf), ('random_forest', rf_model)], voting = 'hard')

# fit the model on training set
ensemble_model.fit(X_train_ss, y_train)

# make predictions on testing set
y_test_pred = ensemble_model.predict(X_test_ss)

# calculate the accuracy
accuracy = accuracy_score(y_test, y_test_pred)
print("Ensemble Method Accuracy:", accuracy)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# initialize the Logistic Regression classifier
lg_clf = LogisticRegression(C = 0.5)

# initialize the feature selection
feature_selection = SelectFromModel(lg_clf)

# fit the model on training set
feature_selection.fit(X_train_ss, y_train)

# select significant features
selected_features = feature_selection.get_support(indices = True)

# convert data to selected features
X_train_selected = feature_selection.transform(X_train_ss)
X_test_selected = feature_selection.transform(X_test_ss)

# initialize the new Logistic Regression classifier
lg_clf_new = LogisticRegression(C = 0.5)

# fit the model on training set with selected features
lg_clf_new.fit(X_train_selected, y_train)

# make predictions on testing set
y_test_pred = lg_clf_new.predict(X_test_selected)

# calculate the accuracy
accuracy = accuracy_score(y_test, y_test_pred)
print("Feature Selection Accuracy:", accuracy)
```

A. Unfortunately, there were no model accuracy increasing compare with the optimal logistic regression classifier with C = 0.5. I think adding more data or cross-validation could be good ways to improving the accuracy.
